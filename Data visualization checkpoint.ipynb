```python
import pandas as pd 

data = pd.read_csv("Video games.csv")
data
```

```python
data.isnull().sum()
```

```python
columns_with_nan = data.columns[data.isnull().any()]

cleaned_data = data.dropna()

cleaned_data.to_csv("cleaned_data.csv", index=False )

cleaned_data
```

```python
cleaned_data.isnull().sum()
```

```python
import matplotlib.pyplot as plt


years = cleaned_data['Year']
global_sales = cleaned_data['Global_Sales']

# Create an area chart
plt.figure(figsize=(10, 6))
plt.stackplot(years, global_sales, labels=['Global Sales'], alpha=0.7)
plt.plot(years, global_sales, marker='o', color='b')  # Add markers and lines for better visibility

# Customize the chart
plt.title('Global Sales Trend Over the Years')
plt.xlabel('Years')
plt.ylabel('Global Sales')
plt.legend(loc='upper left')
plt.grid(True)

# Show the chart
plt.tight_layout()
plt.show()
```

```python
global_sales = cleaned_data['Global_Sales']

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(global_sales, bins=20, color='skyblue', edgecolor='black')

# Customize the chart
plt.title('Distribution of Global Sales')
plt.xlabel('Global Sales')
plt.ylabel('Frequency')
plt.grid(True)

# Show the chart
plt.tight_layout()
plt.show()
```

```python
na_sales = cleaned_data['NA_Sales']
eu_sales = cleaned_data['EU_Sales']

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(na_sales, eu_sales, color='blue', alpha=0.5)

# Customize the chart
plt.title('Relationship between North American Sales and European Sales')
plt.xlabel('North American Sales')
plt.ylabel('European Sales')
plt.grid(True)

# Show the chart
plt.tight_layout()
plt.show()
```

```python
jp_sales = cleaned_data['JP_Sales']
other_sales = cleaned_data['Other_Sales']

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(jp_sales, other_sales, color='green', alpha=0.5)

# Customize the chart
plt.title('Relationship between Japanese Sales and Other Sales')
plt.xlabel('Japanese Sales')
plt.ylabel('Other Sales')
plt.grid(True)

# Show the chart
plt.tight_layout()
plt.show()
```

```python
platform_sales_by_year = cleaned_data.groupby(['Platform', 'Year'])[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum().reset_index()

# Pivot the data to have years as columns and sales by region as values
platform_sales_pivot = platform_sales_by_year.pivot(index='Platform', columns='Year', values=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'])

# Create a stacked area chart
plt.figure(figsize=(12, 8))
platform_sales_pivot.plot(kind='area', stacked=True)

# Customize the chart
plt.title('Distribution of Sales by Platform Over the Years')
plt.xlabel('Platform')
plt.ylabel('Sales')
plt.legend(title='Year')
plt.grid(True)

# Show the chart
plt.tight_layout()
plt.show()
```

```python
publisher_sales = cleaned_data.groupby('Publisher')['Global_Sales'].sum()

# Get the top 10 publishers by global sales
top_publishers = publisher_sales.nlargest(10)

# Create a bar chart
plt.figure(figsize=(10, 6))
top_publishers.plot(kind='bar', color='skyblue')

# Customize the chart
plt.title('Top 10 Publishers by Global Sales')
plt.xlabel('Publisher')
plt.ylabel('Global Sales')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(axis='y')

# Show the chart
plt.tight_layout()
plt.show()
```

```python
# Group the data by genre and calculate total global sales for each
genre_sales = cleaned_data.groupby('Genre')['Global_Sales'].sum()

# Create a pie chart
plt.figure(figsize=(10, 6))
plt.pie(genre_sales, labels=genre_sales.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

# Customize the chart
plt.title('Distribution of Genres by Global Sales')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular

# Show the chart
plt.tight_layout()
plt.show()
```

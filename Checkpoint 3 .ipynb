```python
import pandas as pd


df = pd.read_csv("kc_house_data.csv")
df
```

```python
df.isnull().sum()
```

```python

# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'], format='%Y%m%dT%H%M%S')

# Create a new column with the date part only
df['formatted_date'] = df['date'].dt.date

# Print the updated DataFrame
df
```

```python
df['long'] = df['long'].astype(str)
df['long'] = df['long'].str.replace('-', '')
df['long'] = df['long'].astype(float)
df
```

```python
df.drop('formatted_date' , axis = 1 , inplace = True)
df
```

```python
df.drop('zipcode' , axis = 1 , inplace = True)
df
```

```python
df.drop('date' , axis = 1 , inplace = True)
df.drop('id' , axis = 1 , inplace = True)
df
```

```python
from sklearn.model_selection import train_test_split 

y=df['price']
x=df.drop(columns=['price'])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)

print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the linear regression model
model = LinearRegression()

x_train_reshaped = x_train['grade'].values.reshape(-1, 1)
x_test_reshaped = x_test['grade'].values.reshape(-1, 1)


# Fit the model to the training data
model.fit(x_train_reshaped, y_train)

# Make predictions on the testing data
y_pred = model.predict(x_test_reshaped)

# Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
